```python
import seaborn as sns
import matplotlib.pyplot as plt

##o seaborn está com bug com o colab (atualização) ae precisa do set
sns.set()

plt.figure(figsize=(8,6))
ax = sns.countplot(x = 'composto', data=dados.query('composto in @cod_compostos'),
              order = dat_order['composto'].value_counts().index
              )##.set_title('Top 5 Compostos') é uma forma ou armazena em 'ax'

ax.set_title('Top 5 Compostos em ordem decrescente', fontsize = 20)

plt.xlabel("Compostos",fontsize=16)
plt.ylabel("Quantidade", fontsize=16)

plt.show()
```

Desafio 06: Fazer outras análises com o boxplot e até com o histograma

```python

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10,5))
fig.suptitle('Distribuição de Genicas',fontdict= {'size':20}, weight='bold')
fig. tight_layout(pad=3.0)

ax1.set_title('g0')
ax1.hist(dados['g0'], bins = 50)
ax2.set_title('g19')
ax2.hist(dados['g19'], bins = 50)

ax1.xaxis.set_label_text('Compostos',fontdict= {'size':14}, weight='bold')
ax1.yaxis.set_label_text('Quantidade',fontdict= {'size':14}, weight='bold')

ax2.xaxis.set_label_text('Compostos',fontdict= {'size':14}, weight='bold')
ax2.yaxis.set_label_text('Quantidade',fontdict= {'size':14}, weight='bold')

plt.show()

```
